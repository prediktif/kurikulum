### 1. Pandas - Basic

```python
import numpy as np
import pandas as pd
```

1. Pandas memiliki dua struktur data, yaitu series dan dataframe.
2. Series merupakan kumpulan data berbentuk array yang memiliki index (1 dimensi).
3. Dataframe merupakan kumpulan data berbentuk tabel yang memiliki index dan kolom (2 dimensi).
4. Series dapat dibentuk dari list ataupun array dimensi 1.
5. Dataframe dapat dibentuk dari list, dictionary ataupun array dimensi 2.

```python
contoh_list = [1, 3, 5, 6, 7, 8]
s = pd.Series(contoh_list)
s
```

```python
np.array(contoh_list)
```

```python
pd.Series(np.array(contoh_list))
```

```python
contoh_dict = {
    '0' : [1, 3, 5, 6, 7, 8]
}
pd.DataFrame(contoh_dict)
```

```python
pd.DataFrame(np.array(contoh_list).reshape(-1,1))
```

### 3. Pandas - Conditional Selection in DataFrame

```python
df[df['Sex'] == 'male'].head()
```

```python
df[df['Age'] > 20].head()
```

```python
df['Embarked'].value_counts()
```

```python
df[df['Embarked'].isin(['S', 'Q'])].head()
```

```python
df[df['Embarked'].isna()].head()
```

```python
df[~df['Embarked'].isna()].head()
```

```python
df[df['Name'].str.contains('Mr. ')].head()
```

```python
df[(df['Name'].str.contains('Mr. ')) & (df['Age'].isnull())].head()
```

```python
df[
    (df['Name'].str.contains('Mr. ')) & 
    (df['Age'].isnull()) & 
    (df['Fare'] > 10)
].head()
```

### 4. Pandas - Data Processing

```python
df.info()
```

```python
df['Embarked'].value_counts()
```

```python
df[df['Embarked'].isna()]
```

```python
# Fill null value with other category
df['Embarked'].fillna('other', inplace=True)
```

```python
df[df['Embarked'].isna()]
```

```python
df['Embarked'].value_counts()
```

```python
df[df['Age'].isna()]
```

```python
df['Age'].mean()
```

```python
# Fill null value with mean of age
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df[df['Age'].isna()]
```

```python
# extract title information and create new column
df['Title'] = df['Name'].str.split(' ').str[1]
```

```python
df.head()
```

```python
df['Pclass'].value_counts()
```

```python
# Convert class to be category
df['Pclass'] = df['Pclass'].map({1: 'Economy', 2: 'Business', 3:'Luxury'})
```

```python
df['Pclass'].value_counts()
```

```python
df.head()
```

```python
df['Family Number'] = df['SibSp'] + df['Parch']
```
